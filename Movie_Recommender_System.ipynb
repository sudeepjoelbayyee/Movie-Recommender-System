## First we need to Bring...
- [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
- ['Action','Adventure','Fantasy','"Science Fiction']


```python
import ast
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
import ast
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

## Problem

- Now we need to concatenate all the lists in overview,genres,keywords,cast,crew columns and convert to a string.
- But the Problem is that if there are two words like "James Cameron", It will be problematic, So we need to apply transformation and change to "JamesCameron".
- This is important for recommender system because if there are two seperate words like "James Cameron", there might be a case where there is another name like "James john". Both James are different but as they are treated same as there is space, we get recommendations of "James John" when asked for "James Cameron".
- So If we remove space like "JamesCameron" and "JamesJohn". Now whole name is one entity rather than James seperately and Other name seperately.

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

## Now we need to do text vectorization. Before that we will do lemmatization

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
from sentence_transformers import SentenceTransformer
```

```python
# Prepare the text data for Word2Vec
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
sentences = new_df['tags'].apply(lambda x: [word for word in x.split() if word not in stop_words])
```

```python
# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    words = nltk.word_tokenize(text)
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return " ".join(lemmatized_words)

# Apply lemmatization to 'tags' column
new_df['tags'] = new_df['tags'].apply(lemmatize_text)
```

```python
new_df['tags'][0]
```

## Vectorization

```python
from gensim.models import Word2Vec
```

```python
# Train the Word2Vec model
model = Word2Vec(sentences=sentences, vector_size=5000, window=5, min_count=1, workers=4)
```

```python
model.wv
```

```python
# Function to get sentence vectors (unchanged)
def get_sentence_vector(sentence):
    vectors = [model.wv[word] for word in sentence if word in model.wv]
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)
```

```python
# Get vector representation for each tag
sentence_vectors = [get_sentence_vector(sentence) for sentence in sentences]
```

```python
vectors = np.array(sentence_vectors)
```

## Lets now calculate Cosine similarities

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
new_df.iloc[2662]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-mpnet-base-v2')
new_df = new_df.reset_index(drop=True)
new_df
```

```python
sentence_embeddings = model.encode(new_df['tags'],)
```

```python
sentence_embeddings.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(sentence_embeddings)
```

- The first row similarity matrix has the similarity of first movie with all other movies.
- That is why the first value in this is ~1 which is the similarity of first movie and itself.
- Likewise we have similarity of all movies with the first movie, second movie and so on..

```python
similarity[0]
```

```python
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
